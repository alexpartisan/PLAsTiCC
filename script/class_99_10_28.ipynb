```python
import pandas as pd
import numpy as np
```

```python
ans = pd.read_csv(r'../result/10_28.csv')
```

```python
meta = pd.read_csv(r'../input/test_set_metadata.csv')
meta.head()
```

```python
# Hard coding class_99 of Galactic and ExtraGalactic to constants 0.017 and 0.17 improved my scores --Giba
meta['in_gal'] = 0
meta.loc[meta.distmod.isnull(), 'in_gal'] = 1
meta.in_gal.describe()
```

```python
meta['class_99'] = 0.17
meta.loc[meta.distmod.isnull(), 'class_99'] = 0.017
class_99 = meta[['object_id','class_99']]
```

```python
# class_99.to_pickle('../feat/class_99_test_set.pkl')
```

```python
del(ans['class_99'])
ans.head()
```

```python
ans = ans.merge(class_99, on='object_id', how='left')
```

```python
cols = ans.columns.tolist()
classes = cols[1:]
```

```python
ans[classes].astype(np.float64, inplace=True)
```

```python
ans['object_id'] = ans['object_id'].astype(np.int64)
```
